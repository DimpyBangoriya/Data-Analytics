```python
# attrition vs experience
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(data=hr[hr['Attrition'] == 'Yes'], x='YearsAtCompany', shade=True, label='Attrition = Yes', color='red')
sns.kdeplot(data=hr[hr['Attrition'] == 'No'], x='YearsAtCompany', shade=True, label='Attrition = No', color='blue')

plt.title("KDE Plot of Years at Company vs. Attrition", fontsize=14)
plt.xlabel("Years at Company", fontsize=12)
plt.ylabel("Density", fontsize=12)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(data=hr[hr['Attrition'] == 'Yes'], x='PerformanceRating', shade=True, label='Attrition = Yes', color='red')
sns.kdeplot(data=hr[hr['Attrition'] == 'No'], x='PerformanceRating', shade=True, label='Attrition = No', color='blue')

plt.title("PerformanceRating vs. Attrition", fontsize=14)
plt.xlabel("PerformanceRating", fontsize=12)
plt.ylabel("Density", fontsize=12)
plt.legend()
plt.show()
```
