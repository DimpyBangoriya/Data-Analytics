```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# Titanic Project Insights
```

```python
td=pd.read_csv('Class Data 1 titanic W9 L1.csv')

```

```python
td.head()
```

```python
td.info()
```

```python
td.drop(columns=['Cabin'],inplace=True)
```

```python
# charts with seaborn
```

```python
# Q1. Calculate total number of passenger in each passenger class
```

```python
sns.countplot(data=td,x='Pclass',color='red')
plt.title('Total Passenger in Each Class')
plt.xlabel('Passenger Class')
plt.ylabel('Total Passenger')


```

```python
# Q2. Calculate total fare spending by each class of passenger
```

```python
sns.barplot(data=td,x='Pclass',y='Fare')
plt.title('Passenger Class vs Fare Spent')
plt.xlabel('Passenger Class')
plt.ylabel('Fare Spent')
```

```python
# Q3. Calculate how many survived by gender?
```

```python
sns.countplot(data=td,x='Survived',hue='Sex')


plt.ylabel('Total Passenger')
```

```python
# Q4. Calculate fare spent by gender and also cantegorize by survived and non-survicved?
```

```python
sns.barplot(data=td,x='Sex',y='Fare',hue='Survived')

plt.ylabel('Fare Spent')
```

```python
# Q5.  Find total fare spent by gender in each passenger class
```

```python
sns.barplot(data=td,x='Pclass',y='Fare',hue='Sex')

plt.xlabel('Passenger Class')
plt.ylabel('Fare Spent')
```

```python
# Q6.  Calculate survived and death count in different age groups
```

```python
# create a new column with age group
```

```python
td['Age_Group']=pd.cut(td['Age'],bins=[0,18,30,45,60,80,100])
```

```python
td.head()
```

```python
sns.countplot(data=td,x='Age_Group',hue='Survived')

plt.ylabel('Total Passenger')
```

```python
# Q7. Using Pie chart calculate ratio of percentage of survival and death 

```

```python
v1 = td['Survived'].value_counts()
```

```python
v1
```

```python
plt.pie(v1.values,v1.index,autopct='%0.1f%%')
plt.axis('equal')

```
